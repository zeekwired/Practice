```python
# Import modules
import pandas as pd 
import numpy as np
```

```python
# Read dataset
df = pd.read_csv(r'C:\Users\zachd\OneDrive\Documents\GitHub\Practice\Week 2\bikeshare_train - bikeshare_train.csv', encoding='unicode_escape')
df.head()
```

#### 1. Import the data the drop the 'casual' and 'registered' columns. These are redundant with your target, 'count'.

```python
# Dropping columns: 'casual' and 'registered'
df.drop(columns=['casual', 'registered'], inplace=True)
df.head()
```

#### 2. Transform the 'datetime' column into a datetime type and use it to create 3 new columns in the data frame containing the:

```python
# Converting 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'], format=r'%Y-%m-%d %H:%M:%S')
df.info()
```

1. Name of the Month

```python
df['Month'] = df['datetime'].dt.month_name()
```

2. Name of the Day of the Week

```python
df['Day of the Week'] = df['datetime'].dt.day_name()
```

3. Hour of the Day

```python
df['Hour of day'] = df['datetime'].dt.hour
df.head()
```

    1. Make sure all 3 new columns are 'object' datatype so they can be one-hot encoded later. 

```python
df.info()
```

```python
df['Hour of day'] = df['Hour of day'].astype('str')
df['Hour of day'].dtype
df.info()
```

    - 2. Drop the 'datetime' and 'season' columns. These are now redundant.

```python
df.drop(columns=['datetime', 'season'], inplace=True)
df.head()
```

### 3. The temperatures in the 'temp' and 'atemp' columns are in Celsius. Use `.apply()` to convert them to Fahrenheit using a Lambda function.

```python
df[['temp', 'atemp']] = df[['temp', 'atemp']].apply(lambda x: (x * 9/5) + 32)
df.head()
```

#### 4. Create a new column, 'temp_variance' which shows how much warmer or colder the current temp ('temp') is than the average temp ('atemp').  If the current temperature is warmer than average, the value in 'temp_variance' should be positive. 

```python
df['temp_variance'] = df[['temp', 'atemp']].apply(lambda x: df['temp'] if x['temp'] > x['atemp'] else x['temp'] - x['atemp'], axis=1)
df.head()
```
